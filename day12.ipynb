```python
orig = []
# open file as reading
with open('input1.txt', 'r') as f:
    # read
   
    for line in f:
        orig.append([i for i in line.strip()])

orig
```

```python
from collections import defaultdict
directions = [(1,0),(-1,0),(0,1),(0,-1)]
visited = [[False for i in range(len(orig[0]))] for j in range(len(orig))]
area_map = defaultdict(int)
def bfs(i,j):
    
    queue = collections.deque()
    queue.append((i,j,"n"))
    
    curr = orig[i][j]
    area = 1
    perimeter = 0
    up = []
    down = []
    left = []
    right = []
    
    
    while queue:
        i,j,prev_dir = queue.popleft()
        for x,y in directions:
            new_i,new_j = i+x,j+y
            valid_i = 0 <= new_i < len(orig)
            valid_j = 0 <= new_j < len(orig[0])
            new_dir = ""
            if x == 1:
                new_dir = "v"
            elif x == -1:
                new_dir = "^"
            elif y == 1:
                new_dir = ">"
            elif y == -1:
                new_dir = "<"
            if valid_i and valid_j:
                if not visited[new_i][new_j]: 
                    if orig[new_i][new_j] == curr:
                        visited[new_i][new_j] = True
                        queue.append((new_i,new_j,new_dir))
                        area += 1
                if orig[new_i][new_j] != curr:
                    perimeter += 1   
                    if new_dir == "v":
                        down.append((i,j))
                    elif new_dir == "^":
                        up.append((i,j))
                    elif new_dir == ">":
                        right.append((i,j))
                    elif new_dir == "<":
                        left.append((i,j))
            else:
                perimeter += 1
                if new_dir == "v":
                    down.append((i,j))
                elif new_dir == "^":
                    up.append((i,j))
                elif new_dir == ">":
                    right.append((i,j))
                elif new_dir == "<":
                    left.append((i,j))
    
    
    # print(curr, area, perimeter)
    up.sort()
    down.sort()
    
    # sort left on the second index 
    left = sorted(left, key=lambda x: (x[1], x[0]))
    right = sorted(right, key=lambda x: (x[1], x[0]))
    # print("up" , up)
    # print("down", down)
    # print("left", left)
    # print("right", right)

    new_para = 0
    if up:
        new_para += 1
        prev = up[0]
        for i in range(1,len(up)):
            each = up[i]
            if prev[0] == each[0] and prev[1] + 1 == each[1]:
                prev = each
                continue
            prev = each
            new_para += 1

        # print("up",new_para)
    if down:  
        prev = down[0]
        new_para += 1
        for i in range(1,len(down)):
            each = down[i]
            if prev[0] == each[0] and prev[1] + 1 == each[1]:
                prev = each
                continue
            prev = each
            new_para += 1
        # print("down",new_para)
    
    if left:
        prev = left[0]
        new_para += 1
        
        for i in range(1,len(left)):
            each = left[i]
            if prev[1] == each[1] and prev[0] + 1 == each[0]:
                prev = each
                continue
            prev = each
            new_para += 1
        # print("left",new_para)
    # for i 
    
    if right:
        prev = right[0]
        new_para += 1
        
        for i in range(1,len(right)):
            each = right[i]
            if prev[1] == each[1] and prev[0] + 1 == each[0]:
                prev = each
                continue
            prev = each
            new_para += 1
        # print("right",new_para)
    # for direction, new_i,new_j in perimeter_set:
    #     new_directions = []
        
    #     if direction == "v" or "^":
    #         new_directions = [(0,1)]
    #     else:
    #         new_directions = [(1,0)]
        
        
    #     for x,y in new_directions:
    #         new_new_i,new_new_j = new_i+x,new_j+y
    #         if (direction, new_new_i,new_new_j) in perimeter_set:
    #             print("old", direction,new_i,new_j   )
    #             print("new", direction, new_new_i,new_new_j)
    #             perimeter -= 1  
            

    # print(curr,area,perimeter)
    
    
    return area * new_para

price = 0
for i in range(len(orig)):
    for j in range(len(orig[0])):
        # print(i,j)
        if not visited[i][j]:
            visited[i][j] = True
            price += bfs(i,j)


print(price)    
```
