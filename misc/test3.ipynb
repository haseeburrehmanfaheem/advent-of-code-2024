```python
import collections
directions = [(1,0),(-1,0),(0,1),(0,-1)]


def bfs(i,j):
    
    queue = collections.deque()
    queue.append((i,j,0))
    reachable = 0
    while queue:
        i,j,curr_elev = queue.popleft()
        for x,y in directions:
            new_i,new_j = i+x,j+y
            
            valid_i = 0 <= new_i < len(orig)
            valid_j = 0 <= new_j < len(orig[0])
            if valid_i and valid_j:
                    next_elev = orig[new_i][new_j]
                    
                    if curr_elev + 1 == next_elev:
                        if next_elev == 9:
                            reachable += 1
                        else:
                            queue.append((new_i,new_j,next_elev))
    
    return reachable

```

```python
with open('input.txt', 'r') as f:
    lines = f.readlines()

X_MAX = 101
Y_MAX = 103

pos_array = []
for each in lines:
    pos_unformatted = each.strip().split(" ")[0].split("=")[1].split(",")
    vel_unformatted = each.strip().split(" ")[1].split("=")[1].split(",")
    
    pos_x = int(pos_unformatted[0])
    pos_y = int(pos_unformatted[1])
    
    vel_x = int(vel_unformatted[0])
    vel_y = int(vel_unformatted[1])
    
    pos_array.append((pos_x, pos_y, vel_x, vel_y))

# Optimize neighbor search with a set
def find_neighbors(pos_set, x, y):
    return sum(
        (nx, ny) in pos_set
        for nx, ny in [(x + 1, y), (x - 1, y), (x, y + 1), (x, y - 1)]
    )

max_found = 0
for curr_sec in range(30000):
    # Use set for fast position lookups
    pos_set = set()
    new_pos_array = []

    # Update positions and wrap coordinates
    for pos_x, pos_y, vel_x, vel_y in pos_array:
        pos_x = (pos_x + vel_x) % X_MAX
        pos_y = (pos_y + vel_y) % Y_MAX
        new_pos_array.append((pos_x, pos_y, vel_x, vel_y))
        pos_set.add((pos_x, pos_y))
    
    pos_array = new_pos_array

    # Count neighbors
    near_found = sum(find_neighbors(pos_set, x, y) for x, y, _, _ in pos_array)

    max_found = max(max_found, near_found)

    if near_found > 900:  # Print grid if condition met
        print(curr_sec)
        grid = [["." for _ in range(Y_MAX)] for _ in range(X_MAX)]
        for x, y, _, _ in pos_array:
            grid[x][y] = "#"
        print("\n".join("".join(row) for row in grid))
        print()

print(max_found)
```
