```python
import copy
matrix = []
with open('input.txt', 'r') as file:
    file_content = file.read()
    
    for each in file_content.split("\n"):
        tmp = []
        for char in each:
            tmp.append(char)
        matrix.append(tmp)
        
print(matrix)

antinodes = set()

for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        
        if matrix[i][j] != ".":
            curr_char = matrix[i][j]
            
            for new_i in range(len(matrix)):
                for new_j in range(len(matrix[0])):
                    if new_i == i and new_j == j:
                        continue
                    
                    # print(i,j,new_i,new_j, len(matrix), len(matrix[0]))
                    if matrix[new_i][new_j] == curr_char:
                        antinodes.add((i,j))
                    
                        antinodes.add((new_i,new_j))
                        
                        # print(i,j,new_i,new_j)
                        diff_i = new_i - i
                        diff_j = new_j - j
                        
                        tmp1_i = new_i 
                        tmp1_j = new_j
                        while tmp1_i + diff_i < len(matrix) and tmp1_i + diff_i >= 0 and tmp1_j + diff_j < len(matrix[0]) and tmp1_j + diff_j >= 0:
                            # print(tmp1_i + diff_i,tmp1_j + diff_j)
                            # if matrix[tmp1_i + diff_i][tmp1_j + diff_j] == curr_char:
                            #     continue
                            antinodes.add((tmp1_i + diff_i, tmp1_j + diff_j))
                            tmp1_i += diff_i
                            tmp1_j += diff_j
                        
                        tmp2_i = i
                        tmp2_j = j
                        while tmp2_i - diff_i < len(matrix) and tmp2_i - diff_i >= 0 and tmp2_j - diff_j < len(matrix[0]) and tmp2_j - diff_j >= 0:
                            # print(tmp2_i - diff_i,tmp2_j - diff_j)
                            # if matrix[tmp2_i - diff_i][j - diff_j] == curr_char:
                            #     continue
                            antinodes.add((tmp2_i - diff_i, tmp2_j - diff_j))
                            tmp2_i -= diff_i
                            tmp2_j -= diff_j
                        
                        tmp1_i = new_i 
                        tmp1_j = new_j
                        while tmp1_i - diff_i < len(matrix) and tmp1_i - diff_i >= 0 and tmp1_j - diff_j < len(matrix[0]) and tmp1_j - diff_j >= 0:
                            # print("new", tmp1_i - diff_i,tmp1_j - diff_j)
                            # if matrix[tmp1_i - diff_i][tmp1_j - diff_j] == curr_char:
                            #     continue
                            antinodes.add((tmp1_i - diff_i, tmp1_j - diff_j))
                            tmp1_i -= diff_i
                            tmp1_j -= diff_j
                        
                        tmp2_i = i
                        tmp2_j = j
                        while tmp2_i + diff_i < len(matrix) and tmp2_i + diff_i >= 0 and tmp2_j + diff_j < len(matrix[0]) and tmp2_j + diff_j >= 0:
                            # print("new2", tmp2_i + diff_i,tmp2_j + diff_j)
                            # if matrix[i + diff_i][j + diff_j] == curr_char:
                            #     continue
                            antinodes.add((tmp2_i + diff_i, tmp2_j + diff_j))
                            tmp2_i += diff_i
                            tmp2_j += diff_j

print(antinodes)
print(len(antinodes))
                    
            
        
```
